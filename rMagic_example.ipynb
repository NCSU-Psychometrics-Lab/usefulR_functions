```python
import pandas as pd 
```

```python
df = pd.DataFrame({'Letter': ['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'c'], 
                   'X': [4, 3, 5, 2, 1, 7, 7, 5, 9], 
                   'Y': [0, 4, 3, 6, 7, 10, 11, 9, 13], 
                   'Z': [1, 2, 3, 1, 2, 3, 1, 2, 3]})
```

```python
df.head(5)
```

```python
%%R -i df
ggplot(data = df) + 
    geom_point(aes(x = X, y = Y, color = Letter, size = Z))
```
